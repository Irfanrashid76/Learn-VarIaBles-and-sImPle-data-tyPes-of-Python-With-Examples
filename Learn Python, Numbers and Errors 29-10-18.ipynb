### Avoiding Type Errors with the str() Function

#### Type error: It means Python can’t recognize the kind of informa- tion you’re using.

```python
age = "23"
message = "Happy" + " " + age  + "rd Birthday"

print(message)
```

### String, str() function

```python
age = "23"
message = "Happy" + " " + str(age)  + "rd Birthday"

print(message)
```
